### _AQ7 :Extract tweet source from `source` column
### Code


```python
# check for the unique values
clean_twt_archive.source.unique()
```

```python
#make a function fix_source which extract the strings between tags
def fix_source(x):
    i= x.find('>') + 1
    j =x[i:].find('<')
    # extract the text in between
    return x[i:][:j]
```

```python
clean_twt_archive.source = clean_twt_archive.source.apply(lambda x: fix_source(x)).astype('category')
```

## 1: Comparing dog stages ;

```python
data = twitter_archive_wrangle.groupby('dog_stage').count()['tweet_id']
ax = sns.barplot(y=data.index, x=data.values, palette='Blues_d')
ax.set(xlabel='Count', ylabel='Dog stage', title='Dog Stage Counts')
plt.show()
```

## * Insights.
### As we can see .. 
### 1.Pupper has the highest percentage
### 2. Floofer has the lowest percentage


## 2: The relation between 'retweet_count' and 'favorite_count' ;

```python
# Scatter plot to show the relation between favorits and retweets
ax = sns.regplot(x='retweet_count', y='favorite_count', data=twitter_archive_wrangle, fit_reg=False) 
ax.set(xlabel='Retweet count', ylabel='Favorite count', title='Favorits VS Retweets')
plt.show()
```

```python
ax = sns.regplot(x='retweet_count', y='favorite_count', data=twitter_archive_wrangle, color='b', scatter_kws={'s':5, 'alpha':.3}) 
ax.set(xlabel='Retweet count', ylabel='Favorite count', title='Favorits VS Retweets')
plt.show()
```

## ** Insights.
### As we can see .. 
### A liner relationship between retweet count and favorite count


## 3: Comparing tweets from different sources ;

```python
data = twitter_archive_wrangle.groupby('source').count()['tweet_id']
ax = sns.barplot(y=data.index, x=data.values, palette='Blues_d')
ax.set(xlabel='Count', ylabel='Tweet source', title='Tweet Source Counts')
plt.show()
```

## *** Insights.
